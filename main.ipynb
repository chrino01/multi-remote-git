```python
import numpy as np
import matplotlib.pyplot as plt
```

GENERATE DATA


```python
jum_data = 100
y=[i*0.1+np.random.randn() for i in range(jum_data)]
x=[i*0.1 for i in range(jum_data)]
```

PLOT DATA


```python
plt.scatter(x,y)
plt.show()
```

Membuat fungsi linear

```python
def linear(x,gradien):
    y=gradien*x
    return y
```

```python
# Menghitung pesamaan lniear awal
x_prediction=np.array([0,10])
m_prediction=1
y_prediction= linear(x_prediction,m_prediction)
```

```python
plt.scatter(x,y)
plt.plot(x_prediction,y_prediction,"Red")
plt.axis([0,10,0,10])
plt.show()
```
